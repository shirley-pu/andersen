```python
from mpl_toolkits import mplot3d
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

u0 = np.linspace(-3, 3, 21)
c0 = np.linspace(-0.3, 0.3, 21)
U0, C0 = np.meshgrid(u0, c0)

# H coefficients are those for the relaxed boundary condition, C18:1 lipid, and default parameter values
# three orders of magnitude (same as the parameter values) 
Hb=69.6
Hx=-161
Hc=-42.2

# Eq.7 from manuscript
def f(u,c):
    return (Hb*(u**2))+(Hx*u*c)+(Hc*(c**2))
Gdef_RB = f(U0, C0)

fig=plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(U0, C0, Gdef_RB, 50, cmap='binary')
ax.set_xlabel('u0')
ax.set_ylabel('c0')
ax.set_zlabel('Gdef_RB');
#it may be nice to have the zero Gdef_RB plane indicated?
```

```python
ax1 = plt.axes(projection='3d')
ax1.plot_surface(U0, C0, Gdef_RB, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax1.set_title('surface')
ax1.set_xlabel('u0')
ax1.set_ylabel('c0')
ax1.set_zlabel('Gdef_RB');
```
